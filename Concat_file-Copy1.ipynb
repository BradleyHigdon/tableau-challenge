```python
import pandas as pd
```

```python
# File to Load
file_path = "Resources/JC-202001-citibike-tripdata.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame
file_df = pd.read_csv(file_path, encoding="utf-8")
file_df
```

```python
# File to Load
file_path2 = "Resources/JC-202003-citibike-tripdata.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame
file_df2 = pd.read_csv(file_path2, encoding="utf-8")

file_df2
```

```python
combined_df = pd.concat([file_df, file_df2], axis=0, join='outer', ignore_index=False, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True)
combined_df
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv') 
combined_df.to_csv('out.zip', index=False,
          compression=compression_opts)  
```
